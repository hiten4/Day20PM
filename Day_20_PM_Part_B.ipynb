```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

normal_data = np.random.normal(loc=50, scale=10, size=1000)
standard_normal = (normal_data - np.mean(normal_data)) / np.std(normal_data)


# Plot comparison
plt.figure()

plt.hist(normal_data, bins=30, alpha=0.6, label="Normal (μ=50, σ=10)")
plt.hist(standard_normal, bins=30, alpha=0.6, label="Standard Normal (μ=0, σ=1)")

plt.title("Normal vs Standard Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```


# 2. Hypothesis testing for two groups

```python
group1 = np.random.normal(50, 10, 100)
group2 = np.random.normal(55, 10, 100)

mean1 = np.mean(group1)
mean2 = np.mean(group2)

diff = mean1 - mean2

print("Mean Group1:", mean1)
print("Mean Group2:", mean2)
print("Difference in means:", diff)

if diff != 0:
    print("Groups show difference in average values")
else:
    print("No difference")
```
